### Inherited Limitations

This dataset is well put together and the only missing values are in the extra files provided. Unfortunately this doesn't mean it is ready for an ML pipeline yet. The biggest issue with the dataset is the most interesting fields/features are ones determined by spotify. We have little to no insight into the methods they used but [from the descriptions of said features](https://developer.spotify.com/documentation/web-api/reference/#object-trackobject) we know that the fields like "loudness", "liveness", and "energy" ect are confidence scores which suggests a ML/NN solution on the raw song information. While this is helpful in the sense that we don't have to process thousands of songs to get these features, it does hinder our goal because we don't know what was used to build these features. 

As an example the feature "valence" is defined as a score detailing how 'happy' or 'upbeat' a song is, 1 being the most happy and 0 being the least. The problem arises from the lack of insight into the derivation of this term. Did Spotify use the song's tempo score among other features to determine the 'positivity' value? If these are composite features or contain other features we run the risk of data leakage / contamination since one of these features might be constructed from the value we are attempting to predict.

### Extra Data Files

The main file in this kaggle repository is "data.csv" and this will be our primary dataset. This file is a list of all individual tracks and their audio features, as determined by spotify. There are however several other files provided by the author featuring some different grouping methods.

 The block quotes are the dataset author's explination of each file. 
- data
>Each row represents a single track, each column represents a field of the track (audio features and identifiers)
- data_by_artist
>This file contains the audio features of each artist, resulted from the aggregation. The rows represent different artists, columns represent different audio features. (The original data.csv file is grouped by artists of the tracks, and this csv file is the result of the aggregation process.)

- data_by_genres
>This file contains the audio features of each genre. The rows represent different genres, and the columns represent different audio features.
(The list of genres associated with each artist are added to the "data_by_artist.csv", and the dataset is grouped by the respective genres. As the result of aggregation process, this data file is created.)
- data_by_year
>This data file contains the track data grouped by the year of release of each track, and allows time-series operations to be performed. Each row represents a single year, each column represents an audio feature. The bias of data is as minimal as possible.
- data_with_genre 
>This file is an extension to the "data_by_artist.csv" file with genres implementation for each artist. Each row represents a single artist, each column represents an audio feature.
This data is essential for accomplishing the Recommendation Engine Task if you have decided to build genre-based recommender system.

As you can see there was a significant amount of thought and effort put into the creation of this dataset and the deliniation of what is the author's work and what is Spotify's adds to the author's credibility.

```python
# pandas doing the heavy lifting 
my_correlation = df.corr()

# hiding the upper half of the triangle since it is a repeat anyway
mask = np.triu(np.ones_like(my_correlation,dtype=bool))

#making axis and grid
fig, ax = plt.subplots(figsize=(11,9))

# diverging color map
cmap = sns.diverging_palette(230, 20, as_cmap=True)
corr_mat = sns.heatmap(my_correlation, mask=mask, cmap=cmap, 
            center=0, square=True, linewidths=0.5)
```
